```python
def readable_timedelta(days):
    """gives us readable time in form of weeks/days rather than only days"""
    if days>=7:
        week=str(days//7)
        day=str(days%7)
        return(week+"week(s) and "+day+"day(s)")
    else:
        return("0 week(s) and "+str(days)+" day(s)")
```

```python
# test your function
print(readable_timedelta(10))
```

```python
numbers = [
              [34, 63, 88, 71, 29],
              [90, 78, 51, 27, 45],
              [63, 37, 85, 46, 22],
              [51, 22, 34, 11, 18]
           ]
averages = list(map(lambda num_list:sum(num_list)/len(num_list),numbers))
print(averages)
```

```python
cities = ["New York City", "Los Angeles", "Chicago", "Mountain View", "Denver", "Boston"]

def is_short(name):
    return len(name) < 10

short_cities = list(filter(lambda name:len(name)<10, cities))
print(short_cities)
```
